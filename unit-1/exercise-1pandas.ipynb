```python
import numpy as np
marks=np.array([[78,85,90],
                  [45,60,55],
                  [88,92,95],
                  [30,40,35],
                  [65,70,75]])
print("matrix shape:-")
print(marks.shape)
print("number of elements:-")
print(marks.size)
print("matrix array datatype:-")
print(marks.dtype)
print("student2 all subject marks:-")
print(marks[1:2])
print("subject3 marks of all students:-")
print(marks[:,2])
print("student1 subject 1:- ",marks[:1,1:2])
sum=np.sum(marks,axis=1)
subject=np.sum(marks,axis=0)
sum//=3
subject//=5
print("avergae greater than 80 students:-")
print(np.where(sum>=80,sum,0))
print("grace marks:-")
print(marks+5)
print("pass/fail mappping:-")
print(np.where(marks>49,1,0))
avg = np.average(marks,axis=1)
print("grades allocation")
print(np.where(avg>84,"Score:10",np.where(avg>69,"Score: 8",np.where(avg>49,"Score 6","Score :0"))))
backup=marks
backup1=marks.copy()
marks[0,0]=0
print("checking the copy vs view:-")
print("assignment:-")
print(backup)
print("copy function array:-")
print(backup1)
```
