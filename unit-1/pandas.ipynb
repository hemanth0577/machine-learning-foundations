```python
import numpy as np
price=np.array([1500,3000,5000])
sales=np.array([[10,15,8,12],
                [5,7,4,6],
                [2,3,2,4]])
print(price.shape)
print(price.ndim)
print(sales.shape)
print(sales.ndim)
sales+2
#print(np.dot(price,sales))
print(np.sqrt(sales))
print("sales of smartphone")
print(sales[0:1])
print("sales of all products mon to tues:-")
print(sales[:][0:2])
print("speaker sales on thursday:-",sales[1,3])
print("sales of headphones and smartphones for monday to tuesday:-")
print(sales[1:,0:2])
print(sales.reshape(1,12))
print(sales.reshape(2,6))
print(np.linspace(0,100,10))
print(np.random.randint(0,20,size=(3,4)))
print(np.random.rand(5))
print("sales total:-",sales.sum())
print("sales total row wise:-",sales.sum(axis=1))
```
