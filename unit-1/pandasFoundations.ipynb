```python
import pandas as pd
s=pd.Series([10,20,30,40])
print(s)
print("index at 0:-",s[0])
c=pd.Series([10,20,30], index=["a","b","c"])#custom indexing
print(c)
print("index at custom a is:-",c["a"])
data={"math":90,"physics":85,"chemistry":88}#dictonary
marks=pd.Series(data)
print(data)
print(marks)
```

```python
data={
    "name":["Aman","krish","karan"],
    "age":[22,23,21],
    "marks": [85,90,88]
}
print(data)
print(type(data))
marks=pd.DataFrame(data)
print(marks) #pandas automatically adda a default index: this is called a rangeindex(starts from 0 to n-1)
```

```python
marks=pd.DataFrame(
    data,
    index=["student1","student2","student3"] #custom indexing in pandas data frame
)
print(marks)
marks.set_index("name",inplace=True) #if we are setting inplace true it means we are changing the orginial object if we do this without
print(marks)
```

```python
df=pd.DataFrame ({
     "name":["Aman","krish","karan"],
     "age":[22,23,21],
     "marks": [85,90,88]
}
 )
print(df)
print(df.head())
print(df.tail())
```

```python
df=pd.DataFrame({
    "name":["Aman","saurav","karan","shubham","asbhiek","sammer"],
    "age":[22,23,21,22,25,26],
    "marks": [85,90,88,90,56,70]
})
print(df.head(2))
print(df.tail(2))
print(df.shape)#columns and rows
df.info()# information about the datsets in it
print(df.columns)#colmuns names in the datasets
print(df.describe()) #giving percentile and standard deviation
print(df["name"])#selecting data of colmuns to print
df[["name","marks"]]#printing multiple colmuns data
```

Filtering

```python
print(df[df["marks"]>85])# inner df gives the marks data then taking before outer df only take the data which satisfy the conditins
df[(df["age"]>21)&(df["marks"]>85)]#multiple condition
```

```python
#sorting by default ascending order
print(df.sort_values("marks"))#based on passed colmn name it data will be sorted
df.sort_values("age",ascending=False)#to print int he descending order
```

```python
df["age"].value_counts()#print the how many times does the data repeated in the passed colmun
```

```python
#missing data concept
#nan is for only numerical datatype
df=pd.DataFrame({
    "name":["Aman","saurav","karan","shubham",None,"sammer"],
    "age":[22,23,None,22,25,26],
    "marks": [85,None,88,90,56,None]
})
print(df)
print(df.isnull())#prints true in  missing values
print(df.isnull().sum())#prints the count of null values for each columns
print(df.dropna())#for upfate the real data we have to use the inplace true or store into new dataframe and use it
cleaned_data=df.dropna()
print(cleaned_data)
```
