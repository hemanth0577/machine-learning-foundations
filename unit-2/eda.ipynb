```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data ={
    "Name":["Aman","Riya","Rahul","Sneha","Arjun","Priya","Karan","Neha"],
    "Age":[22,25,24,23,26,27,28,22],
    "Gender":["Male","Female","Male","Female","Male","Female","Male","Female"],

    "Marks":[75,88,92,70,85,90,78,95]
}
df=pd.DataFrame(data)
```

```python
plt.plot(df["Marks"])

plt.title("Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
plt.bar(df["Name"], df["Marks"])

plt.title("Marks Of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")

plt.show()
```

# histogram used for divide data into group of class calcalute bin maxdata minus mindata and interval size will be the bin divided by the mean

```python
plt.hist(df["Age"], bins=5)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

# used for the regression problem scatter graph

```python
plt.scatter(df["Age"], df["Marks"])

plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")

plt.show()
```

#pie chart

```python
gender_counts = df["Gender"].value_counts()

plt.pie(gender_counts,labels=gender_counts.index, autopct="%1.1f%%")

plt.title("Gender Distribution")

plt.show()
```

#explianing the grid with the colors

```python
plt.bar(df["Name"], df["Marks"], color="green")

plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)#shows the two graphs in figure block first parameter is rows and second is the colmuns and last will represent in which qudarant the graphs will present
plt.hist(df["Age"])
plt.title("Age Distribution")

plt.subplot(1,2,2)
plt.bar(df["Name"], df["Marks"])
plt.title("Marks of Students")

plt.show()
```
